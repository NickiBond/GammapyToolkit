```python
import numpy as np

def log_bin_edges_to_centers(edges):
    """
    Convert logarithmically spaced bin edges to bin centers.
    """
    edges = np.asarray(edges)
    if np.any(edges <= 0):
        raise ValueError("All edges must be positive for logarithmic bins.")
    
    centers = np.sqrt(edges[:-1] * edges[1:])
    return centers

# Example usage:
# edges = np.logspace(0, 2, num=5)  # 1, 3.16, 10, 31.6, 100

edges = [0.126,
    0.398,
    1.259,
    3.981,
    12.589,
    39.811,
    126.893]
centers = log_bin_edges_to_centers(edges)
print("Edges:", edges)
print("Centers:", centers)
```
