```python
def CompareCSVToNewRunlist(csv_path, new_runlist_path, missing_file_path = None):
    # Compare working csv to new runlist

    df = pd.read_csv(csv_path)
    df=df[df['include']== 'yes']
    Current = set(df['run_id'])


    New = set(np.loadtxt(new_runlist_path, dtype=str))
    New = set(int(i) for i in New)

    print("Runs in Current but not in New:")
    print(sorted(Current.difference(New)))
    print("\nRuns in New but not in Current:")
    print(sorted(New.difference(Current)))

    if missing_file_path is not None:
        with open(missing_file_path, 'w') as f:
            for run in sorted(New.difference(Current)):
                f.write(f"{run}\n")
```
